using System;

namespace TriangleAreaCalculator
{
    class Program
    {
        static void Main(string[] args)
        {
            double sidea = 2;
            double sideb = 2;
            double sidec = 2;

            double area = TriangleArea(sidea, sideb, sidec);
            Console.WriteLine($"The area of a triangle with sides {sidea}, {sideb}, and {sidec} is: {area}");
        }

        static double TriangleArea(double a, double b, double c)
        {
            double s = (a + b + c) / 2;
            double area = Math.Sqrt(s * (s - a) * (s - b) * (s - c));
            return area;
        }
    }
}

```python
using System;

namespace OddEvenSeparator
{
    class Program
    {
        static void Main(string[] args)
        {
            Console.Write("Input the number of elements to be stored in the array: ");
            int numElements = Convert.ToInt32(Console.ReadLine());

            int[] inputArray = new int[numElements];
            int[] evenArray = new int[numElements];
            int[] oddArray = new int[numElements];
            int evenCount = 0, oddCount = 0;

            Console.WriteLine("Input {0} elements in the array:", numElements);
            for (int i = 0; i < numElements; i++)
            {
                Console.Write($"element - {i} : ");
                inputArray[i] = Convert.ToInt32(Console.ReadLine());
                if (inputArray[i] % 2 == 0)
                {
                    evenArray[evenCount] = inputArray[i];
                    evenCount++;
                }
                else
                {
                    oddArray[oddCount] = inputArray[i];
                    oddCount++;
                }
            }

            Console.WriteLine("The Even elements are:");
            for (int i = 0; i < evenCount; i++)
            {
                Console.Write(evenArray[i] + " ");
            }

            Console.WriteLine("\nThe Odd elements are:");
            for (int i = 0; i < oddCount; i++)
            {
                Console.Write(oddArray[i] + " ");
            }
        }
    }
}
```

```python
using System;

namespace RectangleChecker
{
    class Program
    {
        static void Main(string[] args)
        {
            /// Here we test the inside() function
            Console.WriteLine(inside(1, 1, 0, 0, 2, 3)); 
            Console.WriteLine(inside(-1, -1, 0, 0, 2, 3)); 

            /// Here we test the expression for both rectangles
            bool inFirstRectangle = inside(1, 1, 0.3, 0.5, 1.1, 0.7);
            bool inSecondRectangle = inside(1, 1, 0.5, 0.2, 1.1, 2);

            Console.WriteLine($"Does Point (1,1) inside the first rectangle: {inFirstRectangle}");
            Console.WriteLine($"Does Point (1,1) inside the second rectangle: {inSecondRectangle}");
        }

        static bool inside(double x, double y, double x1, double y1, double x2, double y2)
        {
            return (x >= x1 && x <= x2) && (y >= y1 && y <= y2);
        }
    }
}
```

```python
def pig(word):
    vowels = ['a', 'e', 'i', 'o', 'u']
    word = word.lower()
    if word[0] in vowels:
        return word + 'way'
    else:
        return word[1:] + word[0] + 'ay'

print(pig('happy')) 
print(pig('Enter'))  
```

```python
def bldcount(namefile):
    blood_count = {'A': 0, 'B': 0, 'AB': 0, 'O': 0, 'OO': 0}

    with open(namefile, 'r') as file:
        content = file.read()
        blood_types = content.split()
        for blood_type in blood_types:
            blood_count[blood_type] += 1

    for blood_type, count in blood_count.items():
        if count == 0:
            print(f"There are no patients of blood type {blood_type}")
        elif count == 1:
            print(f"There is one patient of blood type {blood_type}.")
        else:
            print(f"There are {count} patients of blood type {blood_type}.")

# We will call the function with the file path where you have saved the file in the computer.
bldcount(r"C:\Users\junej\Downloads\bloodtype1.txt")
```

# Question 10

```python
def frequencies(txt):
    letters = 'abcdefghijklmnopqrstuvwxyz'
    result = [0] * 26
    for char in txt:
        if char.lower() in letters:
            index = letters.index(char.lower())
            result[index] += 1
    return result

print(frequencies('The quick red fox got bored and went home.'))
print(frequencies('apple'))
```
